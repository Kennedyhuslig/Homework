```python
import matplotlib.pyplot as plt 
marks={'Andy':88, 'Amy':66, 'James':90, 'Jules':55, 'Arthur':77}
students=list(marks.keys())
grades=list(marks.values())
plt.bar(students, grades)
plt.title("Grades Bar Chart")
plt.xlabel("Student Name")
plt.ylabel("Grade")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,4,100)
plt.plot(x,4/4*x+8, color='green')
plt.plot(x,3/4*x+5, color='orange')
plt.plot(x,3/4*x+1, color='blue')
plt.ylim(0,13)
plt.xlim(0,4)
plt.xlabel('X Label')
plt.ylabel('Y Label')

plt.show()
```

```python
import matplotlib.pyplot as plt
marks={'Andy':88, 'Amy':66, 'James':90, 'Jules':55, 'Arthur':77}
students=list(marks.keys())
grades=list(marks.values())
plt.pie(grades, labels=students)
plt.legend()
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt

x=np.arange(0,10,0.1)
y1=np.sin(x)
y2=np.cos(x)
y3=[10,20,30,40,50]
labels=['A', 'B','C','D','E']
y4=[20,30,10,40,50]
y5=np.random.rand(100)
y6=np.random.rand(1000)

fig, axs=plt.subplots(2,3,figsize=(12,8))
axs[0,0].plot(x,y1,label='sin(x)')
axs[0,0].plot(x,y2,label='cos(x)')
axs[0,0].set_title("Multiple Lines Plot")
axs[0,0].legend()

axs[0,1].bar(labels,y3)
axs[0,1].set_title("Bar Chart")

axs[0,2].pie(y4,labels=labels)
axs[0,2].set_title("Pie Chart")

axs[1,0].plot(x,y1)
axs[1,0].grid(True)
axs[1,0].set_title("Grid Plot")

axs[1,1].scatter(x[:len(y5)], y5)
axs[1,1].set_title("Scatter Plot")

axs[1,2].hist(y6, bins=30)
axs[1,2].set_title("Histogram Plot")

plt.show()                 
    


              
                                
```

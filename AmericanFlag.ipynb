```python
from turtle import *
setup(800, 500)
speed(0)

def draw_stripe(x, y):
    penup()
    goto(x, y)
    pendown()
    color("red")
    begin_fill()
    for i in range(2):
        forward(800)
        right(90)
        forward(38)
        right(90)
    end_fill()

def draw_stripes():
    stripe_x = -400
    stripe_y = 200
    for i in range(7):
        draw_stripe(stripe_x, stripe_y)
        stripe_y -= 76

def draw_rectangle():
    penup()
    goto(-400, 250)
    pendown()
    color("blue")
    begin_fill()
    for i in range(2):
        forward(350)
        right(90)
        forward(266)
        right(90)
    end_fill()

def draw_star(x, y):
    penup()
    goto(x, y)
    pendown()
    color("white")
    begin_fill()
    for i in range(5):
        forward(20)
        right(144)
    end_fill()

def draw_stars():
    x_start = -380
    y_start = 230
    for row in range(5):
        if row % 2 == 0:
            x = x_start
        else:
            x = x_start + 28.5
        y = y_start - row * 52
        for i in range(6):
            draw_star(x, y)
            x += 57

def draw_flag():
    draw_stripes()
    draw_rectangle()
    draw_stars()
    hideturtle()

draw_flag()
```
